# Decomposition of Time Series

### Trend is long term movement of series,i.e. increasing and decreasing.
### In time series data, seasonality is the presence of variations that occur at specific regular intervals less than a year, such as weekly, monthly, or quarterly. Seasonality may be caused by various factors, such as weather, vacation, and holidays[1] and consists of periodic, repetitive, and generally regular and predictable patterns in the levels of a time series.
### The error or variability associated with the series is the Irregular component. This is also called white-noise.

### Yt: time series value (actual data) at period t.

### St: seasonal component (index) at period t.

### Tt: trend cycle component at period t.

### It: irregular (remainder) component at period t

## There are two types of model
### 1)Additive model: Observation = Trend + Seasonality + Error , Yt= Tt+ St+ It

### 2)Multiplicative model: Observation = Trend * Seasonality * Error ,Yt= Tt* St* It

```python
from statsmodels.tsa.seasonal import seasonal_decompose 
from pylab import rcParams
```

```python
#1) 
decomposition=seasonal_decompose(df['Number of Car Sold'],model='additive')
rcParams['figure.figsize']=15,8 # it resize all the plots to same size you mention.
decomposition.plot();
```

### If the graphs amplitude is constant as the year passes the we can use the additive. Which is flase in above case
### Here Residual plot follows some pattern after decompostion. Means there might be some information that could be explained so it will not be consider as white noice.

```python
#2) 
decomposition=seasonal_decompose(df['Number of Car Sold'],model='multiplicative')
rcParams['figure.figsize']=15,8 # it resize all the plots to same size you mention.
decomposition.plot();
```

### We the graph shows the increasing or decreasing amplitude as year passes by then we can use the multiplicative model 
### Here we can see the Residual graph at last is constant through out the year passes by so we can say that all the information were explained by the multiplicative model.
