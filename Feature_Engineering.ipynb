```python
import pandas as pd
```

```python
data=pd.read_csv('daily-total-female-births-CA.csv',parse_dates=[0],header=0)
```

```python
data.head()
```

## feature engineering

### 1) Date-Time Feature
### 2) Lag Feature
### 3) Window Feature
### 4) Expanding Feature

# 1) Data-Time Feature

```python
feature= data.copy()
```

```python
feature['year']= data['date'].dt.year
```

```python
feature['month']= data['date'].dt.month
```

```python
feature['day']= data['date'].dt.day
```

```python
feature.head()
```

# 2) Lag feature

```python
feature['lag1']=data['births'].shift(1)
```

```python
feature['lag2']=data['births'].shift(362)
```

```python
feature
```

# 3) Window Feature

```python
feature['Roll_mean']=data['births'].rolling(window=7).mean()
```

```python
feature.head(10)
```

```python
feature['Roll_max']=data['births'].rolling(window=2).max()
```

```python
feature.head(10)
```

# 4) Expanding Feature

```python
feature['Exp_mean']=data['births'].expanding().mean()
```

```python
feature.head(10)
```
